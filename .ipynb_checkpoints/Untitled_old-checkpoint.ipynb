```python
import numpy
import matplotlib.pyplot as plt
import pandas
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
numpy.random.seed(7)
```

```python
df = pandas.read_csv('TotalSelectedBirdValuesFinal.csv', usecols=[1,3], engine='python', skipfooter=3)

dataframe = df[df["ID"] == 120]
dataframe = dataframe.drop('ID', 1)
#dataframe = dataframe["latitude":"longitude"]
#dataframe = pandas.read_csv('latplotallbirds.csv', usecols=[1], engine='python', skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
train_size = int(len(dataset) * 0.77)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
print(len(train), len(test))
```

```python
dataset
```

```python
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0:2]
        dataX.append(a[0])
        b = dataset[i + look_back, 0:2]
        dataY.append(b)
    return numpy.array(dataX), numpy.array(dataY)
```

```python
def create_dataset(dataset, look_back=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-look_back-1):
		a = dataset[i:(i+look_back), 0]
		dataX.append(a)
		dataY.append(dataset[i + look_back, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
x,y = create_dataset(dataset)
```

```python
x.shape
```

```python
y.shape
```

```python
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
trainX
```

```python
trainY
```

```python
trainX = numpy.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = numpy.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

```python
#trainY = numpy.reshape(trainY, (trainY.shape[0],))
#trainY = trainY[:][0]
```

```python
def smallmodel():
    model = Sequential()
    model.add(LSTM(4, input_dim=1))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
model = smallmodel()
model.fit(trainX, trainY, nb_epoch=8, batch_size=1, verbose=2)
#model.save("bird_AvgLongitude_model.h5")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

n = [1,2,3,4,5,6,7]
trainingRMSE = [0.14,0.10,0.11,0.08,0.05,0.04,0.04]
CrossValidation = [0.45,0.40,0.41,0.37,0.33,0.36,0.40]
x = [5,5]
y = [0,0.33]
x1 = [0,5]
y1 = [0.33,0.33]

plt.plot(n, trainingRMSE, 'r-',label = 'Training RMSE')
plt.plot(n, CrossValidation, 'g-',label = 'Test RMSE')
plt.plot(x,y,'y--')
plt.plot(x1,y1,'y--')
#plt.plot(n, dct['lst_0.05'], 'b-',label = 'ErrorRate 5%')
#plt.plot(n, dct['lst_0.1'], 'c-',label = 'ErrorRate 10%')
red_patch = mpatches.Patch(color='red',label = 'Training RMSE')
green_patch = mpatches.Patch(color='green',label = 'Test RMSE')
#blue_patch = mpatches.Patch(color='blue',label = 'ErrorRate 5%')
#cyan_patch = mpatches.Patch(color='cyan',label = 'ErrorRate 10%')
plt.legend(loc='upper right',handles=[red_patch,green_patch])
plt.xlabel("No of Epochs")
plt.ylabel("RMSE 3 fold Cross Validation")
plt.axis([0, 8, 0, 1])
plt.show()
```
